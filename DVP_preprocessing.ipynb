```python
import pandas as pd
import numpy as np

csv_file =  pd.read_csv("dutch_data/dvpdb/questions.csv")


def print_full(x):
    pd.set_option('display.max_rows', len(x))
    print(x)
    pd.reset_option('display.max_rows')


duplicates = csv_file.groupby(by="QuestionDuplicateState")
questions_by_duplicates = duplicates.size()
print(questions_by_duplicates)

for key, item in duplicates:
    print(duplicates.get_group(key), "\n\n")
```
