```python
import pandas as pd
questions = pd.read_csv("dutch_data/dvpdb/questions.csv")

#questions = questions["Question"]

print(questions)
```

```python
import csv
with open("dutch_data/dvp_formatted.csv", "w") as file:
    csv_file = csv.writer(file)
    csv_file.writerow(["question1", "question2"])
    
    index = 0
    for index, row in questions.iterrows():
        question = row["Question"]
        question1 = ""
        question2 = ""
        if index % 2 == 0:
            question1 = question
        else:
            question2 = question
            csv_file.writerow([question1, question2])
        index = index + 1

    
```
